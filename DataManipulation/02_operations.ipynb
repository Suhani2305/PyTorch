```python
import torch as tc
X= tc.arange(20).reshape(4,5)
X
```

```python
# exp: it returns a new tensor with the exponential of the elements of the input tensor
exp_X = tc.exp(X)
exp_X
```

```python
# cat using dim =0 (vertically) and dim=1 (horizontally)
cat_dim0 = tc.cat([X, exp_X], dim=0)
cat_dim1 = tc.cat([X, exp_X], dim=1)
cat_dim0, cat_dim1
```

```python
#  ones - creates a tensor filled with ones
ones_tensor = tc.ones((2, 5) )
zeros_tensor = tc.zeros((2, 5) )
ones_tensor == zeros_tensor   # element-wise comparison
```

```python
ones_tensor.sum()   # sum of all elements in the tensor
```

## Broadcasting

Broadcasting works according to the following two-step procedure:

1. **Expand dimensions**  
   One or both arrays are expanded by copying elements along axes with length **1**, so that after this transformation, both tensors have the **same shape**.

2. **Element-wise operation**  
   Once the shapes match, an **element-wise operation** is performed on the resulting arrays.


```python
a = tc.arange(3).reshape(3, 1)
b = tc.arange(2).reshape(1, 2)
a, b
```

```python
a+b, a*b, a-b
```

```python
X= 1
Y= 2
```

```python
Y = X+Y    # new memory is allocated and y kaa adress change ho jata hai
Y
```

```python
# in-place operation
X= 1
Y= 2
Y += X   # no new memory is allocated, y kaa adress change nahi hota
```
