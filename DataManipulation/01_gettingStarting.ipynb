```python
x = tc.arange(12, dtype=tc.float32).reshape(3, 4)
x
```

```python
# numel() - returns the total number of elements in the tensor
x.numel()  # 12 elements
```

```python
#  shape - returns the shape of the tensor
x.shape  # torch.Size([3, 4])
```

```python
#  reshape - changes the shape of the tensor
x_reshaped = x.reshape(4, 3)
x_reshaped
```

```python
# zeros - creates a tensor filled with zeros
zeros_tensor = tc.zeros((2, 5, 3) )
zeros_tensor
```

```python
#  ones - creates a tensor filled with ones
ones_tensor = tc.ones((2, 5) )
ones_tensor
```

```python
#  randn - creates a tensor with random numbers from a normal distribution
randn_tensor = tc.randn((3, 4), dtype=tc.float32)
randn_tensor
```
