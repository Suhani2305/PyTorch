```python
import math
import time
import numpy as np 
import torch as th
# from d2l import torch as d2l
```

```python
n = 10000
a = th.ones(n)
b = th.ones(n)

c = th.zeros(n)
t = time.time()
for i in range(n):
    c[i] = a[i] + b[i]
print('%.5f sec' % (time.time() - t))
```
